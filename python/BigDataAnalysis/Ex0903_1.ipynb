```python
# Pyplot, cv2   

import cv2
import matplotlib.pyplot as plt

#  read  
img1r = cv2.imread('C:/kyj/spyder_python/07DataAnalysis/image1_680448.jpg') # img read
img1 = cv2.cvtColor(img1r, cv2.COLOR_BGR2RGB) # Convert color (BGR-> RGB)
img2r = cv2.imread('C:/kyj/spyder_python/07DataAnalysis/image2_550364.jpg') 
img2 = cv2.cvtColor(img2r, cv2.COLOR_BGR2RGB)

# 
img1s = cv2.resize(img1, dsize=(640,480))
img2s = cv2.resize(img2, dsize=(640,480))

#  suplot 
plt.subplot(121)
plt.axis('off') # x/y  (axis() )
plt.imshow(img1s) # img    

plt.subplot(122)
plt.axis('off')
plt.imshow(img2s)

plt.show()
```

```python
#    / Img writing (to PC)
font = cv2.FONT_HERSHEY_SIMPLEX
num = 100
numstr = 'Num:%d'% num # formatting
cv2.putText(img1, numstr, (10,40), font, 1.1, (255,0,0), 2) # Text (image,text,,,(1:100%),(R,G,B), ) 
cv2.putText(img1, 'Cat', (img1.shape[1]-100,img1.shape[0]-30), font, 1.5, (0,0,255), 3) # .shape[0]:height / .shape[1]:width
plt.axis('off')
plt.imshow(img1)
plt.show()

img1c = cv2.cvtColor(img1, cv2.COLOR_RGB2BGR) # Convert color (RGB-> BGR)
cv2.imwrite('C:/kyj/spyder_python/07DataAnalysis/image1_680448_text.jpg',img1c) #img writing. (,img)
```
