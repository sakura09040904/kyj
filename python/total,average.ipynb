```python
import os
import glob
import pandas as pd

all_books = glob.glob(os.path.join('C:\\kyj\\python', 'score_*.xlsx'))
names=[]
cnt = 0
writer=pd.ExcelWriter('total.xlsx')

for book in all_books:
    all_sheets = pd.read_excel(book, sheet_name=None, index_col=None)
    print(os.path.basename(book))
    total=[]
    mean=[]

    for name,data in all_sheets.items():
        totals = 0
        means = 0
        if(cnt == 0):
            names.append(data['이름']) #리스트로 반환됨.
            cnt += 1
        means = data.mean(axis=1).astype(int)#시리즈로 반환됨.
        means.to_list()
        totals = data.sum(axis=1) #시리즈로 반환됨.
        totals.to_list()
        pd1 = pd.DataFrame()
        pd1['이름'] = names
        pd1['합계'] = totals
        pd1['평균'] = means
        pd1.to_excel(writer)
        writer.save()
#print(pd1)
```
