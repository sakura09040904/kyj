- Matplot library : 데이터 시각화를 위한 모듈
    - figure : 그래프를 그려주기 위한 전체 영역
    - plot : 그래프 자체
 
- plt.plot()
    - plt.plot([x_data],[y_data],'색상,형태'))
        - data가 한개일경우 y_data
        - 'r' - red, 'b' - blue, 'g' - green, ... / 'o' - dot, '^' - triangle, '-' - 점선, 's' - square, ...
        - 여러 개의 선을 그릴 경우 순차적으로 입력 plot(x1,y1,f1,x2,y2,f2,....)
    - plot.axis([x_min, x_max, y_min, y_max])
    - plot.xlable(''), plot.ylable('')
    - plot.show()
    
- plt.figure()
    - figure.add_subplot(행의개수, 열의개수, ploot위치)

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[2,4,6,8],'ro')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.show()
```

```python
import numpy as np

t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--', t,t**2,'bs', t,t**3,'g^')
plt.axis([0,5,0,40])
plt.show
```
